#### Create a function that:

1. Given a string "s", reverse it.

```python
def reverter(s):
    return s[::-1]

reverter("Processamento de Linguagem Natural em Engenharia Biomédica")
```

2. Given a string "s", returns how many "a" and "A" characters are present in it.

```python
def contar_a(s):
    return s.count("a") + s.count("A")

contar_a("Processamento de Linguagem Natural em Engenharia Biomédica")
```

3. Given a string "s", returns the number of vowels there are present in it.

```python
def contar_vogais(s):
    vogais = "aeiouAEIOU"
    total = 0
    for c in s:
        if c in vogais:
            total += 1
    return total

contar_vogais("Processamento de Linguagem Natural em Engenharia Biomédica")
```

4. Given a string "s", convert it into lowercase.

```python
def converter_minusculas(s):
    return s.lower()

converter_minusculas("Processamento de Linguagem Natural em Engenharia Biomédica")
```

5. Given a string "s", convert it into uppercase.

```python
def converter_maiusculas(s):
    return s.upper()

converter_maiusculas("Processamento de Linguagem Natural em Engenharia Biomédica")
```

6. Verifica se uma string é uma capicua.

```python
def capicua(s):
    return s == s[::-1]

capicua("Processamento de Linguagem Natural em Engenharia Biomédica")
```

7. Verifica se 2 strings são balanceadas (2 strings, s1 e s2, estão balanceadas se todos os caracteres de s1 estão presentes em s2).

```python
def balanceadas(s1, s2):
    for c in s1:
        if c not in s2:
            return False
    return True

print(balanceadas("Processamento de Linguagem Natural", "Engenharia Biomédica"))
print(balanceadas("Ee", "Engenharia"))
```

8. Calcula o número de ocorrências de s1 em s2.

```python
def ocorrencias(s1, s2):
    return s2.count(s1)

print(ocorrencias("Processamento de Linguagem Natural", "Engenharia Biomédica"))
print(ocorrencias("bio", "biomédica"))
```

9. Verifica se s1 é anagrama de s2.
    - "listen" e "silent": deve imprimir True;
    - "hello", "world": deve imprimir False.

```python
def anagrama(s1, s2):
    return sorted(s1) == sorted(s2)

print(anagrama("listen", "silent"))
print(anagrama("hello", "world"))
```
